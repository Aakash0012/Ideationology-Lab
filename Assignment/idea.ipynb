# Factorial

```python
def fact(n=5):
    if n==1:
        return 1
    else:
        return n*fact(n-1)
```

```python
fact(n=7)
```

```python
class calci:
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def add(self):
        return self.a + self.b
    
    def sub(self):
        return self.a - self.b
    
    def pro(self):
        return self.a * self.b
    
    def div(self):
        return self.a / self.b
    
    def fact(self, n):
        if n==1:
            return 1
        else:
            return n*fact(n-1)
```

```python
obj = calci(4,2)
obj
```

```python
a = obj.add()
a
```

```python
f = obj.fact(n=5)
f
```

```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.w3schools.com/html/html_tables.asp'
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html5lib')

a = soup.findAll('table', attrs = {'id':'customers'})
ro = a[0].findAll('tr')

for i in ro[0].findAll('th'):
    print(i.text, end='\t')
        
print()
for j in range(1, len(ro)):
    for i in ro[j].findAll('td'):
        print(i.text, end='\t')
    print()
```
